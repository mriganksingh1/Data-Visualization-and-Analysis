```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
titanic=pd.read_csv('train.csv')
```

```python
titanic.head(5)
```

# QUESTIONS FOR ANALYSIS

```python
#Which age group has better chances of survival?
#which social class has better chances of survival?
#which gender has better chances of survival?
```

```python
survived=titanic.loc[titanic['Survived']==1]
died=titanic.loc[titanic['Survived']==0]
```

```python
survived['Age'].describe()
```

```python
#As Max 80 so dividing age groups in 4 parts 
classes_survival=survived['Pclass'].value_counts()
```

```python
sns.set(rc={'figure.figsize':(10,8)})
```

```python
vis1=classes_survival.plot(kind='bar')
vis1.set_xlabel('Pclass')
vis1.set_ylabel('count')
vis1.set_title('Social economic class survival')
vis1.set_xticklabels(vis1.get_xticklabels(),rotation=0)
```

```python
classes_death=died['Pclass'].value_counts()
```

```python
vis2=classes_death.plot(kind='bar')
vis2.set_xlabel('Pclass')
vis2.set_ylabel('count')
vis2.set_title('Socio economic class death')
vis2.set_xticklabels(vis2.get_xticklabels(),rotation=0)
```

```python
gender_survived=survived['Sex'].value_counts()
gender_died=died['Sex'].value_counts()
```

```python
vis3=gender_survived.plot(kind='bar')
vis3.set_xlabel('Gender')
vis3.set_ylabel('Count')
vis3.set_title('Gender most survived')
```

```python
vis4=gender_died.plot(kind='bar')
vis4.set_xlabel('Gender')
vis4.set_ylabel('Count')
vis4.set_title('Gender most died')
```

```python
plt.plot(titanic['Fare'],titanic['Survived'])
plt.xlabel('Fare')
plt.ylabel('Survival')
plt.title('Survival vs Fare')
```
